```python
import sounddevice as sd
from scipy.io.wavfile import write

fs = 44100  
seconds = 4

print("Recording...")
myrecording = sd.rec(int(seconds * fs), samplerate=fs, channels=2)
sd.wait()  # Wait 
write('my_voice.wav', fs, myrecording)  # Save 
print("Recording complete.")
```

```python
from tensorflow.keras.models import load_model
model = load_model('speech.keras')
```

```python
def preprocess_audio(filename):
    # Load the audio file same as training time
    y, sr = librosa.load(filename, duration=3, offset=0.5)
    
    # Extract MFCC features
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40)
    
    # Compute the mean same as in training)
    mfccs_mean = np.mean(mfccs.T, axis=0)
    
    # Reshape to match model input (40 timesteps, 1 feature per timestep)
    return mfccs_mean.reshape(40, 1)

```

```python
my_voice_mfcc = preprocess_audio('my_voice.wav')
```

```python
my_voice_mfcc = np.expand_dims(my_voice_mfcc, axis=0)

# Predict the emotion
predicted_class = model.predict(my_voice_mfcc)

# Map the predicted class index to emotion labels
emotions = ['happy', 'sad', 'angry', 'fear', 'neutral', 'disgust', 'ps']  # Update based on dataset
predicted_emotion = emotions[np.argmax(predicted_class)]

print(f"The predicted emotion is: {predicted_emotion}")
```
